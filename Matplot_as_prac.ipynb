```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x=  np.linspace(0,5,11)
y= x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 7, 1, 6]
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('Y axis')
```

Bar charts

```python
x = ["Math","Science","Hindi","Sanskrit"]
y=[78,89,79,65]
plt.bar(x,y)
plt.title("Score")
plt.xlabel("Subject name")
plt.ylabel("Subject Marks")
```

 # Scatter Plot -Visualizing relationships or patterns between two variables.

```python
x=[5,7,8,8,7,2,3,9]
y=[56,78,98,32,66,77,69,89]
plt.scatter(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
```

# Pie Chart-Showing percentages or proportions.

```python
labels = ['Python','java','c++','c']
sizes =[40,25,20,15]
plt.pie(sizes,labels=labels, autopct="%1.1f%%")
plt.title("Programming Language Popularity")
```

# Histogram -Showing distribution of data (e.g. ages, scores, etc.)

```python
data = [22, 55, 62, 45, 21, 22, 34, 42, 42, 4, 99, 102, 110, 120, 121, 122, 130]

plt.hist(data, bins=5)
plt.title("Histogram Example")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

# Subplot in Matplotlib

```python

x=  np.linspace(0,5,11)
y= x**2
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,y)
```

# object oriented way 

```python
fig = plt.figure()
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.1,0.1,.3,.3]) # overlapping 
axis2.plot(x,y)
```

```python
plt.plot(x,y, color = "pink")
```
